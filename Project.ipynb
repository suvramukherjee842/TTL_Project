```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading the csv file 
ipl=pd.read_csv("ipl_2022_dataset.csv")
```

```python
ipl.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
ipl.head()
```

```python
ipl.isnull().sum()
```

```python
# checking Null Values 
ipl[ipl['Cost IN $ (000)'].isnull()]
```

```python
# fill NaN Value WIth 0 
ipl['Cost IN $ (000)']=ipl['Cost IN $ (000)'].fillna(0)
ipl['COST IN ₹ (CR.)']=ipl['COST IN ₹ (CR.)'].fillna(0)
```

```python
#checking NUll 
ipl.isnull().sum()
```

```python
ipl[ipl['2021 Squad'].isnull()]
```

```python
ipl['2021 Squad']=ipl['2021 Squad'].fillna('Not Participated')
```

```python
ipl[ipl['2021 Squad'].isnull()]
```

```python
ipl.isnull().sum()
```

```python
# GEting TEam name Which is greater than 0 crore 
teams=ipl[ipl['COST IN ₹ (CR.)']>0]['Team'].unique()
teams
```

```python
#put Sold Where player Associated with a team
ipl['status']=ipl['Team'].replace(teams,'sold')
```

```python
plt.figure(figsize=(10,5))
fig=sns.countplot(ipl,x='status',palette=['orange','pink'])
plt.xlabel('Sold or Unsold')
plt.ylabel("Number of players")
plt.title("sold vs Unsold",fontsize=16)
fig.bar_label(fig.containers[0])

plt.show()
```
